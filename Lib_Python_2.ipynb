```python
t= np.linspace(0, 10, 51)
t
```

```python
f= np.cos(t)
f
```

```python
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.plot(t,f, label='f=cos(t)', color='green')
title_dict = {'fontsize' : 14, 'fontweight': 'bold', 'color':'lightgrey', 'family': 'Calibri'}
label_dict = {'fontsize' : 12, 'fontweight': 'bold', 'color':'lightgrey', 'family': 'Calibri'}
plt.xlabel('Переменная t', fontdict= label_dict)
plt.ylabel('Функция f(t)', fontdict=label_dict)
plt.title('График f(t)', fontsize = 14, fontweight = 'bold')
plt.legend(frameon=False)
plt.grid(color='ghostwhite')
plt.show()
```

```python
x= np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x,y1, label='y = x**2', color='green')
ax1.set_title('Функция y = x**2', fontdict=title_dict)
ax1.set_xlim(-5,5)

ax2.plot(x,y2, label='y= 2*x + 0.5', color='red')
ax2.set_title('Функция y= 2*x + 0.5', fontdict=title_dict)

ax3.plot(x,y3, label='y= -3*x - 1.5', color='grey')
ax3.set_title('Функция y= -3*x - 1.5', fontdict=title_dict)

ax4.plot(x,y4, label='y= sin(x)', color='blue')
ax4.set_title('Функция y= sin(x)', fontdict=title_dict)

fig.set_size_inches(8,6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
